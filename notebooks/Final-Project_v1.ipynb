```python
from os import listdir
from os.path import isfile, isdir, join

def get_list_of_files(data_path):
    # get list of files from the data path
    files = [file for file in listdir(data_path) if isfile(join(data_path, file))]
    # get list of folders from the data path
    dirs = [(file, join(data_path, file)) for file in listdir(data_path) if isdir(join(data_path, file))]
    # get files from each folder
    for i, dir_ in enumerate([d[1] for d in dirs]):
        files += [(dirs[i][0], join(dir_, file), file) for file in listdir(dir_) if isfile(join(dir_, file))]
    return files

data_1 = get_list_of_files('Data_source_1')
data_2 = get_list_of_files('Data_source_2')
print(len(data_2))
data_2[:10]
```

```python
import PyPDF2
import pandas as pd

def get_text_from_pdf(file_path):
    file = open(file_path, 'rb')
    reader = PyPDF2.PdfReader(file)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    file.close()
    return text

def convert_pdfs_into_dataframe(pdfs):
    rows = []
    for pdf in pdfs:
        rows.append({'file': pdf[2], 'profession': pdf[0], 'resume':get_text_from_pdf(pdf[1])})
    return pd.DataFrame(rows)

df = convert_pdfs_into_dataframe(data_2)
df.tail()
```

## Verbs simmilarity analysis

```python
from nltk import word_tokenize
import string
from nltk.corpus import stopwords
from textblob import TextBlob

def clean_text(text):
    text = text.translate(str.maketrans('', '', string.punctuation)) # remove punctuation
    tokens = word_tokenize(text) # tokenize words
    sw = stopwords.words('english') # remove stopwords
    tokens = [w.lower() for w in tokens if w.lower() not in sw]
    tokens = [w for w in tokens if not w.isdigit()] # remove digits/numbers
    return ' '.join(tokens)

def get_tokenized_unique_words(text):
    tokens = word_tokenize(text)
    print("# total words:", len(tokens)) # remove repeated words
    tokens = list(set(tokens))
    print("# unique words:", len(tokens))
    return tokens

def get_pos_tags(tokens, tag_start=''):
    blob = TextBlob(' '.join(tokens)) # pos tagging words
    tags = [tag for tag in blob.tags if tag[1].startswith(tag_start)]
    return tags
```

```python
from nltk.stem import WordNetLemmatizer

def lemmatize_words(tokens, pos):
    lem = WordNetLemmatizer() # lemmatize words
    verbs_lem = [lem.lemmatize(token[0], pos=pos) for token in tokens]
    verbs_lem = get_tokenized_unique_words(' '.join(verbs_lem))
    return verbs_lem

text = df.iloc[2479]['resume']
text = clean_text(text)
tokens = get_tokenized_unique_words(text)
verbs = get_pos_tags(tokens, 'VB')
verbs_resume = lemmatize_words(verbs, 'v')
verbs_resume[:10]
```

```python
text = job_desc
text = clean_text(text)
tokens = get_tokenized_unique_words(text)
verbs = get_pos_tags(tokens, 'VB')
verbs_jobdesc = lemmatize_words(verbs, 'v')
verbs_jobdesc[:10]
```

```python
common_verbs = [verb for verb in verbs_resume if verb.lower() in verbs_jobdesc]

print("% verbs simmilarity: {:.2f}".format(len(common_verbs) * 100 / len(verbs_jobdesc)))
common_verbs
```

## Keywords (named entities) simmilarity analysis

```python
from nltk import sent_tokenize
import nltk

def get_named_entities(text):
    sentences = sent_tokenize(text) # sentence tokenization
    tokens = [word_tokenize(sent) for sent in sentences] # word tokenize each sentence
    tags = [nltk.pos_tag(sent) for sent in tokens] # pos tagging each word
    chunks = nltk.ne_chunk_sents(tags, binary=True) # set named entity chunks
    named_entities = []
    for sent in chunks: # save named entities from text
        for chunk in sent:
            if hasattr(chunk, "label") and chunk.label() == 'NE':
                named_entities.append(chunk)
    return named_entities
```

```python
def extract_specific_named_entities(nes, tag):
    named_entities = []
    # extract only named entities with equal tag
    for ne in nes:
        named_entity = [word[0] for word in ne if word[1] == tag] # compare tags
        named_entity = ' '.join(named_entity) # stringify list
        if named_entity != "":
            named_entities.append(named_entity) # add named entity to list
    return list(set(named_entities)) # return only unique values

text = job_desc # job description text
nes = get_named_entities(text)
nes_jobdesc = extract_specific_named_entities(nes, 'NNP')
nes_jobdesc
```

```python
text = df.iloc[2479]['resume'] # resume text
text = clean_text(text)
tokens = get_tokenized_unique_words(text)
```

```python
common_nes = [word for word in tokens if word.lower() in list(map(str.lower, nes_jobdesc))]

print("% keyword simmilarity: {:.2f}".format(len(common_nes) * 100 / len(nes_jobdesc)))
common_nes
```

## Bigrams simmilarity analysis

```python
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

def clean_text_without_lowering(text):
    text = text.translate(str.maketrans('', '', string.punctuation)) # remove punctuation
    tokens = word_tokenize(text) # tokenize words
    sw = stopwords.words('english') # remove stopwords
    tokens = [w for w in tokens if w.lower() not in sw]
    tokens = [w for w in tokens if not w.isdigit()] # remove digits/numbers
    return ' '.join(tokens)

def get_top_n_bigrams(tokens, n):
    bcf = BigramCollocationFinder.from_words(tokens) # instantiate bigram finder
    bigrams = bcf.nbest(BigramAssocMeasures.likelihood_ratio, n) # get top n bigrams
    return bigrams

text = job_desc # job description text
text = clean_text_without_lowering(text)
bigrams_jobdesc = get_top_n_bigrams(word_tokenize(text), 20)
bigrams_jobdesc
```

```python
text = df.iloc[2479]['resume'] # resume text
text = clean_text_without_lowering(text)
bigrams_resume = get_top_n_bigrams(word_tokenize(text), 1000)
len(bigrams_resume)
bigrams_resume[:10]
```

```python
bigrams_jobdesc_lowered = [tuple(map(str.lower, bigram)) for bigram in bigrams_jobdesc]
bigrams_resume_lowered = [tuple(map(str.lower, bigram)) for bigram in bigrams_resume]
# bigrams_resume_lowered += [('apply', 'passionate')]
bigrams_jobdesc_lowered[:10]
```

```python
common_bigrams = [b for b in bigrams_resume_lowered if b in bigrams_jobdesc_lowered]

print("% bigrams simmilarity: {:.2f}".format(len(common_bigrams) * 100 / len(bigrams_jobdesc_lowered)))
common_bigrams
```

```python
df_concat = pd.concat([df_jobdesc, df_resume], ignore_index=True)
df_concat = df_concat.fillna(0.0)
df_concat = df_concat.iloc[:, : df_jobdesc.shape[1]]
df_concat
```

```python
df_resume_concat = df_concat.iloc[-len(df_resume):]
df_resume_concat.reset_index(drop=True, inplace=True)
df_resume_concat
```

```python
import spicy

matrix_jobdesc = spicy.sparse.csr_matrix(df_jobdesc.values)
matrix_resume = spicy.sparse.csr_matrix(df_resume_concat.values)

print("Job Description matrix shape:", matrix_jobdesc.shape)
print("Resume matrix shape:", matrix_resume.shape)
```

```python
simmilarity_matrix = matrix_jobdesc * matrix_resume.T
simmilarity_matrix
```

```python
simmilarity_matrix.todense() # simmilarities between each resume entry (columns) and each job description entry (rows)
```

```python
best_simmilarities = []

# loop through job description entries and find the resume entry with most simmilarity
for i in range(matrix_jobdesc.shape[0]):
    simmilarities = [matrix.item(i) for matrix in simmilarity_matrix.todense().T] # save simmilarities
    max_sim = max(simmilarities) # get max simmilarity
    max_sim_index = simmilarities.index(max_sim)
    best_simmilarities.append((max_sim_index, max_sim))

best_simmilarities
```

```python
for i in range(matrix_jobdesc.shape[0]):
    print("Job description entry:\t\t", sents_jobdesc[i])
    best_sim = best_simmilarities[i][1]
    print("% Best simmilarity:\t\t", round(best_sim * 100, 2), "%")
    best_sim_index = best_simmilarities[i][0]
    print("Best simmilarity in resume:\t", sents_resume[best_sim_index] if best_sim != 0.0 else "NOT FOUND")
    print("-"*30)
```
